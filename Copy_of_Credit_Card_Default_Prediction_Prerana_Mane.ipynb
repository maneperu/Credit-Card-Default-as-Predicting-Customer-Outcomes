## **Note** :-


In this dataset, the repayment status from April 2005 to September 2005 is represented by the variables PAY_0 to PAY_6. These variables utilize a scale to indicate different payment behaviors:

1. A value of -2 signifies no consumption.
2. -1 indicates that the payment has been made in full.
3. 0 represents the use of revolving credit, where only the minimum payment has been made.
4. Positive values (1 to 9) indicate payment delays, with higher numbers indicating longer delays. For instance, 1 signifies a payment delay of one month, 2 indicates a delay of two months, and so on, up to 8 for a delay of eight months, and 9 for nine months or more.

This scale helps categorize individuals based on their payment patterns over the specified time period, providing valuable insights into their credit behavior.

In our dataset, we have the transaction history of customers' credit card usage for the past six months. Our task is to predict whether a customer will default on their payments or not.

#**SMOTE**



*   Imbalanced classification involves developing predictive models on classification datasets that have a severe class imbalance.



*   The challenge of working with imbalanced datasets is that most machine learning techniques will ignore, and in turn have poor performance on, the minority class, although typically it is performance on the minority class that is most important.


*   One approach to addressing imbalanced datasets is to oversample the minority class. The simplest approach involves duplicating examples in the minority class, although these examples don’t add any new information to the model. Instead, new examples can be synthesized from the existing examples. This is a type of data augmentation for the minority class and is referred to as the Synthetic Minority Oversampling Technique, or **SMOTE** for short.
*  SMOTE (Synthetic Minority Oversampling Technique) – Oversampling is one of the most commonly used oversampling methods to solve the imbalance problem.
It aims to balance class distribution by randomly increasing minority class examples by replicating them







```python
#import the module
from imblearn.over_sampling import SMOTE

smote = SMOTE()

# fit predictor and target variable
x_smote , y_smote = smote.fit_resample(credit_df.iloc[:,0:-1],credit_df['default_payment_next_month'])

print('Original Dataset Shape - ',len(credit_df))
print('Resampled Dataset Shape  - ',len(y_smote))
```

# **Logistic Regression**

```python
#import necessary packages
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score, roc_auc_score, confusion_matrix, roc_curve, auc
```

```python
#make a copy of dataframe
credit_df_logistic = credit_df_copy.copy()
```

```python
credit_df_logistic.head()
```

### train test split data set

```python
#X and y values
X = credit_df_logistic.drop(['default_payment_next_month','total_Payement_Value','Dues'],axis = 1)
y = credit_df_logistic['default_payment_next_month']
```

```python
columns = X.columns
```

```python
#standardise the x value by using satandardscaler
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
#split the dataset
X_train, X_test , y_train , y_test = train_test_split(X, y , test_size = 0.33 , random_state = 42, stratify = y)
```

**What is GridSearchCV?**

GridSearchCV is a function provided by scikit-learn's model_selection package. Its purpose is to systematically explore a specified grid of hyperparameters for a given estimator (model). By doing so, it enables us to find the best combination of hyperparameters for our model by fitting it to the training data. This way, we can optimize our model's performance before deploying it for predictions

```python
#set the parameter
param_grid = {'penalty' :['l1','l2'],'C' : [0.001, 0.01, 0.1, 1, 10, 100, 1000]}
```

```python
#fit the parameter
grid_lr_clf = GridSearchCV(LogisticRegression(), param_grid, scoring = 'accuracy', n_jobs = -1, verbose = 3, cv = 3)
grid_lr_clf.fit(X_train, y_train)
```

```python
optimized_clf = grid_lr_clf.best_estimator_
```

```python
grid_lr_clf.best_params_
```

```python
grid_lr_clf.best_score_
```

```python
# Predicted Probability
train_preds = optimized_clf.predict_proba(X_train)[:,1]
test_preds = optimized_clf.predict_proba(X_test)[:,1]
```

```python
# Get the predicted classes
train_class_preds = optimized_clf.predict(X_train)
test_class_preds = optimized_clf.predict(X_test)
```

## Model Evaluation

```python
# Get the accuracy scores
train_accuracy_lr = accuracy_score(train_class_preds,y_train)
test_accuracy_lr = accuracy_score(test_class_preds,y_test)

print("The accuracy on train data is ", train_accuracy_lr)
print("The accuracy on test data is ", test_accuracy_lr)
```
